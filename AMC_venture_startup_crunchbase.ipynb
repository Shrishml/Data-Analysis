## Making a right aggregation function 

Data is not distributed gaussian. We could have used mean or median as the aggregration function for the pivot table. That might not be appropriate for the case. We made another aggregration function that computes the percentage of data that lies between the range given the the client.

We define the thresold to be 30 percent. Meaning that we will select all those investment type haviing more then 30 percent data in the range we have set.

```python
def agg_func(arr,a=5e6,b=15e6):
    return percentileofscore(arr,b)-percentileofscore(arr,a) # % value
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.percentileofscore.html    
```
